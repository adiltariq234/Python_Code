```python
import tkinter as tk

def check_age():
    age = entry.get().strip()

    if age == "":
        label.config(text="‚ùå Enter your age", fg="red")
    elif not age.isdigit():
        label.config(text="‚ùå Please enter numbers only", fg="red")
    else:
        age = int(age)

        if age < 18:
            label.config(text="üßí You are young", fg="blue")
        elif age < 30:
            label.config(text="üî• You are smart", fg="green")
        else:
            label.config(text="üß† You are experienced", fg="purple")

root=tk.Tk()
root.title("ElifProject")
root.geometry("400x300")
entry=tk.Entry(root,width=40)
entry.pack(pady=10)
btn=tk.Button(root,text="Check Age",command=check_age)
btn.pack(pady=10)
label=tk.Label(root,font=("Arial",18))
label.pack(pady=10)

root.mainloop()
```

```python
import tkinter as tk

def show_table():
    num = entry.get().strip()

    if num == "":
        label.config(text="‚ùå Enter your number", fg="red")
        return

    if not num.isdigit():
        label.config(text="‚ùå Please enter numbers only", fg="red")
        return

    num = int(num)
    result = ""

    for i in range(1, 11):
        result += f"{num} √ó {i} = {num * i}\n"

    label.config(text=result, fg="green")

root = tk.Tk()
root.title("Table Project")
root.geometry("400x300")

entry = tk.Entry(root, width=30)
entry.pack(pady=10)

btn = tk.Button(root, text="Find Table", command=show_table)
btn.pack(pady=10)

label = tk.Label(root, font=("Arial", 14), justify="left")
label.pack(pady=10)

root.mainloop()
```

```python
import tkinter as tk

root = tk.Tk()
root.title("Two Frames")
root.geometry("400x300")

top_frame = tk.Frame(root, bg="white", height=200)
top_frame.pack(fill="both")

bottom_frame = tk.Frame(root, bg="lightgray", height=100)
bottom_frame.pack(fill="x")

tk.Label(top_frame, text="Chat Area", font=("Arial", 16)).pack(pady=60)
tk.Label(bottom_frame, text="Input Area").pack(pady=20)

root.mainloop()
```
