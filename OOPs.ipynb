```python
class House:
  def __init__(self,Color,Size):
    self.Color=Color
    self.Size=Size
  def describe(self):
    print(f"The House Is {self.Color} and {self.Size}")

my_house=House("Black","Large")
my_house.describe()
```

```python
class BankAccount:
  def __init__(self,balance,Name):
    self.balance=int(balance) # Ensure balance is an integer
    self.Name=Name
  def deposit(self,amount):
    self.balance +=amount
    print(amount,"Add ho gaya hai")
  def Withdraw(self, amount):
    if amount <=self.balance:
      self.balance -=amount
      print(amount,"Withdraw ho gaya hai")
    else:
      print("Insufficient Balance")
  def checkbalance(self):
    print("Balanced",self.balance)

acc1=BankAccount(1000,"Adil")
print(acc1.Name)
acc1.deposit(1000)
acc1.checkbalance() #
acc1.Withdraw(500)
acc1.checkbalance()
```
