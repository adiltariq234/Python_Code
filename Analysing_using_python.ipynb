```python
import pandas as pd
import numpy as np

data={
    "A":[1,2,3,4,np.nan],
    "B":["Adil","Adeel","Ayan",np.nan,"Ali"],
    "C":["Cat","Dog","Frog",np.nan,"Snack"],
    "D":[1,2,3,4,5]
}
df=pd.DataFrame(data)
print(df)
```

```python
print("Dummies Variable of with inducuter ")
print(pd.get_dummies(df,columns=["C"]))
```

```python
import pandas as pd

marks=[23,24,25,26,27,45,56,65,87,67,87,45,76,45,76]
name=["Adil","Adeel","Ayan","Ali","Adil","Adeel","Ayan","Ali","Adil","Adeel","Ayan","Ali","Adil","Adeel","Ayan"]

# DataFrame banana
df = pd.DataFrame({
    "Name": name,
    "Marks": marks
})

print("Original form of data")
print(df)

# Binning / Discretization
bins = [0,50,60,80,90,100]
labels = ["Fail","Normal","B","A","A+"]

df["Grade"] = pd.cut(df["Marks"], bins=bins, labels=labels)

print("\nAfter Grading")
print(df)
```
