```python
import pandas as pd

marks=[23,24,25,26,27,45,56,65,87,67,87,45,76,45,76]
name=["Adil","Adeel","Ayan","Ali","Adil","Adeel","Ayan","Ali","Adil","Adeel","Ayan","Ali","Adil","Adeel","Ayan"]

# DataFrame banana
df = pd.DataFrame({
    "Name": name,
    "Marks": marks
})

print("Original form of data")
print(df)

# Binning / Discretization
bins = [0,50,60,80,90,100]
labels = ["Fail","Normal","B","A","A+"]

df["Grade"] = pd.cut(df["Marks"], bins=bins, labels=labels)

print("\nAfter Grading")
print(df)
```

```python
import pandas as pd
import numpy as np

data = {
    "Sales": [12000,13000,14000,15000,16000,10000,13000,123400]
}

index = [
    ("Pakistan","Lahore",2023),
    ("Pakistan","Karachi",2022),
    ("Pakistan","Islamabad",2021),
    ("Pakistan","Lahore",2020),
    ("Pakistan","Karachi",2019),
    ("Pakistan","Islamabad",2018),
    ("Pakistan","Lahore",2017),
    ("Pakistan","Karachi",2016)
]

multi_index = pd.MultiIndex.from_tuples(
    index,
    names=['Country','City','Year']
)

df = pd.DataFrame(data, index=multi_index)
print(df)
```

```python
df.groupby(level="Year").sum()
```
