```python
class Car:
  def __init__(self,brand,color):
    self.brand=brand
    self.color=color
  def start(self):
    print(f"The {self.brand} {self.color} car has started")
  def stop(self):
    print(f"The {self.brand} {self.color} car has stopped")

car1=Car("Toyota","Red")
car2=Car("Honda","Blue")

car1.start()
car2.stop()
```
