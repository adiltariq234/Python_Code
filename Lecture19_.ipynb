```python
class Car:
  def __init__(self,brand,color):
    self.brand=brand
    self.color=color
  def start(self):
    print(f"The {self.brand} {self.color} car has started")
  def stop(self):
    print(f"The {self.brand} {self.color} car has stopped")

car1=Car("Toyota","Red")
car2=Car("Honda","Blue")

car1.start()
car2.stop()
```

```python
from transformers import pipeline

# Load model
pipe = pipeline(
    "text-generation",
    model="koutch/qwen_qwen3-instruct-4b_train_sft_train_code",
    trust_remote_code=True
)

# Take input from user
user_input = input("Enter your question: ")

# Create messages dynamically
messages = [
    {"role": "user", "content": user_input},
]

# Generate response
response = pipe(messages)

# Print output clearly
print("\nModel Response:\n")
print(response[0]["generated_text"])
```

```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="koutch/qwen_qwen3-instruct-4b_train_sft_train_code",
    device_map="auto"
)

while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Chat ended.")
        break

    messages = [{"role": "user", "content": user_input}]
    output = pipe(messages, max_new_tokens=500)

    # ✅ CLEAN OUTPUT
    assistant_reply = output[0]["generated_text"][-1]["content"]
    print("AI:", assistant_reply)
```
