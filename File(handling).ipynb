```python
file=open("Example.txt","w")
file.write("Hello This is a Example file")
file.close()
```

```python
new_line="_________New Line_________\n"
with open("Example.txt","r") as file:
    old_data=file.read()
with open("Example.txt","w") as file:
    file.write(new_line + old_data)
```

```python
new_lines="New Data Here"
with open ("Example.txt","r") as file:
    lines=file.readlines()

lines.insert(2,new_lines)
with open("Example.txt","w") as file:
    file.writelines(lines)
```

```python
with open ("Example.txt","r")as file:
    contant=file.read()
print("File Contant: ",contant)

```

```python
import json
data={
    "Name" : "Adil",
    "Age"  : 20,
    "Skill": ["Python","Excel","DataAnaltics"]
}
with open("Example.json","w") as file:
    json.dump(data,file,indent=4)

with open("Example.json","r")as file:
    Adil=json.load(file)
print(Adil)
print(Adil["Name"])
print(Adil["Skill"][2]) 
```
